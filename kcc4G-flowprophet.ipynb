```python
df = pd.read_excel('data/KCC4Gtest.xlsx',sheetname='kcc4g')
df['ds'] = pd.to_datetime(df['ds'])
df.tail(10)
```

```python
future= d.make_future_dataframe(periods=1,freq='M')
```
